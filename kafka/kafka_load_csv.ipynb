```python
df = pd.read_csv('../craw-data/raw-data/item-data/data0.csv')
```

```python
list_data = df.T.to_dict().values()
```

```python
from kafka import KafkaProducer
from json import dumps
kafka_producer = KafkaProducer(bootstrap_servers='localhost:9092', value_serializer=lambda x: dumps(x).encode('utf-8'))
```

```python
for item in list_data:
    kafka_producer.send('item-shopee', value=item)
```
